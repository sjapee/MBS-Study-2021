```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from  matplotlib.ticker import PercentFormatter
#import bokeh_catplot as bkh
#import bokeh.io
import matplotlib.ticker as FuncFormatter
import matplotlib.ticker as ticker
```

```python
#FOR SFACES
SFthreshdata = pd.read_excel('/misc/data12/sjapee/MoebiusSyndrome/Behavioral_Data/AllTaskFMRIData-OSF.xlsx',
                           "StaticFacesTask", header = [0])
#SFthreshdata.columns
```

```python
#SFaces - by Task Type
indata = SFthreshdata
ControlData = indata[indata.Task == "Control Task"]
EmotionData = indata[indata.Task == "Emotion Task"]
custom_pal = ['skyblue','salmon']
gbox=sns.set_palette(custom_pal)


boxprops = {'edgecolor': 'w', 'linewidth': 3}
lineprops = {'color': 'k', 'linewidth': 3}
boxplot_kwargs = {'boxprops': boxprops, 'medianprops': lineprops,
                  'whiskerprops': lineprops, 'capprops': lineprops,
                  'width': 0.75, 'palette': custom_pal}

stripplot_kwargs = {'linewidth': 2, 'size': 10, 'alpha': 0.9,
                    'palette': custom_pal}

fig = plt.figure(figsize=(10,8))
ax1 = fig.add_subplot(1,2,1)
#gbox, ax = plt.subplots()
gbox1 = sns.boxplot(ax=ax1, x="Emotion",y="Threshold",hue="Group",
             data=ControlData,meanline='True',showfliers = False,  **boxplot_kwargs)
                 
gbox1 = sns.stripplot(x="Emotion",y="Threshold",hue="Group", dodge = True, jitter = True,
                    data=ControlData, ax=gbox1.axes, edgecolor='gray', **stripplot_kwargs)
            
gbox1.legend_.remove()
gbox1 = sns.color_palette("Paired")
gbox1 = sns.set_style("ticks")
gbox1 = sns.despine(offset = 10, trim = False)
plt.gca().set_ylim(0,1)
ax1.yaxis.set_major_formatter(PercentFormatter(1))
ax1.set_xlabel('Control Task')
gbox1 = sns.set_context("poster",font_scale=1.2)

ax2 = fig.add_subplot(1,2,2)
gbox2 = sns.boxplot(ax=ax2, x="Emotion",y="Threshold",hue="Group",
             data=EmotionData,meanline='True',showfliers = False, **boxplot_kwargs)
                 
gbox2 = sns.stripplot(x="Emotion",y="Threshold",hue="Group", dodge = True, jitter = True,
                    data=EmotionData,ax=gbox2.axes, edgecolor='gray', **stripplot_kwargs)
            
gbox2.legend_.remove()
gbox2 = sns.color_palette("Paired")
gbox2 = sns.set_style("ticks")
gbox2 = sns.despine(offset = 10, trim = False)
plt.gca().set_ylim(0,1)
#plt.rc('font', family='serif', serif='Arial')
ax2.yaxis.set_major_formatter(PercentFormatter(1))
ax2.set_xlabel('Emotion Task')
gbox2 = sns.set_context("poster",font_scale=1.2)

ax2.spines['left'].set_visible(False)
ax2.axes.get_yaxis().set_visible(False)

for b in ax1.artists:
    b.set_alpha(0.9)
for b in ax2.artists:
    b.set_alpha(0.9)

hatches = [None, None, None, None]
colors = ['skyblue','salmon', 'skyblue',  'salmon']
a = len(ax1.artists)
#print(a)
for hatch, patch, color in zip(hatches, ax1.artists, colors):
    patch.set_edgecolor('white')
    patch.set_facecolor(color)
    patch.set_hatch(hatch)
for hatch, patch, color in zip(hatches, ax2.artists, colors):
    patch.set_edgecolor('white')
    patch.set_facecolor(color)
    patch.set_hatch(hatch)

plt.show()

```

```python
#FOR IE-FMRI ROI DATA
IEFMRIPSCdata = pd.read_excel('/misc/data12/sjapee/MoebiusSyndrome/Behavioral_Data/AllTaskFMRIData-OSF.xlsx',
                           "TaskFMRIData", header = [0])
#IEFMRIPSCdata.columns

IEFMRIPSCtidydata = pd.melt(IEFMRIPSCdata,
                         value_vars=['rFFA','rpSTS', 'rAMG'],
                         id_vars = ['Group','Scode','Task'],
                        var_name = ['ROI'], value_name = 'PSC')
#IEFMRIPSCtidydata
```

```python
#fmri - by Task Type
indata = IEFMRIPSCtidydata
custom_pal = ['skyblue','skyblue','salmon','salmon']
#custom_pal = ['darkseagreen','khaki']
#custom_pal = ['orchid','thistle']
#custom_pal = ['sandybrown','tan']
lcolor = 'darkgray'
lalpha = 0.9
lw1 = 6
msize1 = 10

strip_custom_pal = ['silver','silver']
strip_custom_pal = [lcolor, lcolor]
#strip_custom_pal = custom_pal
gbox=sns.set_palette(custom_pal)

IDData = indata[indata.Task == "Identity"]
ExprData = indata[indata.Task == "Expression"]

FFAData = indata[indata.ROI == 'rFFA']
STSData = indata[indata.ROI == 'rpSTS']
AMGData = indata[indata.ROI == 'rAMG']

boxprops = {'edgecolor': 'k', 'linewidth': 3}
lineprops = {'color': 'k', 'linewidth': 3}
boxplot_kwargs = {'boxprops': boxprops, 'medianprops': lineprops,
                  'whiskerprops': lineprops, 'capprops': lineprops,
                  'width': 0.75, 'palette': custom_pal}

stripplot_kwargs = {'linewidth': lw1, 'size': 0, 'alpha': lalpha,
                    'palette': strip_custom_pal}

fig = plt.figure(figsize=(27,8))
ax1 = fig.add_subplot(1,3,1)
gbox1 = sns.boxplot(ax=ax1, x="Group",y="PSC",hue="Task",
             data=FFAData,meanline='True',showfliers = False, **boxplot_kwargs)
gbox1 = sns.stripplot(x="Group",y="PSC",hue="Task", dodge = True, jitter = False,
            data=FFAData, ax=gbox1.axes, edgecolor=None, **stripplot_kwargs)

#figure out how to add the gray lines for pairs of points
idx0 = 0; idx1 = 1
locs1 = ax1.get_children()[idx0].get_offsets()
locs2 = ax1.get_children()[idx1].get_offsets()
for i in range(locs1.shape[0]):
    x = [locs1[i, 0], locs2[i, 0]]
    y = [locs1[i, 1], locs2[i, 1]]
    ax1.plot(x, y, color= lcolor, alpha=lalpha, linewidth=lw1, markersize = msize1)
idx0 = 2; idx1 = 3
locs1 = ax1.get_children()[idx0].get_offsets()
locs2 = ax1.get_children()[idx1].get_offsets()
for i in range(locs1.shape[0]):
    x = [locs1[i, 0], locs2[i, 0]]
    y = [locs1[i, 1], locs2[i, 1]]
    ax1.plot(x, y, color= lcolor, alpha=lalpha, linewidth=lw1, markersize = msize1)


gbox1.legend_.remove()
gbox1 = sns.set_style("ticks")
gbox1 = sns.despine(offset = 30, trim = False)
#plt.gca().set_ylim(-0.5,2.5)
#ax1.set_xlabel('Identity Task')
gbox1 = sns.set_context("poster",font_scale=1.2)
ax1.set_ylabel('PSC')

###################
ax2 = fig.add_subplot(1,3,2)
gbox2 = sns.boxplot(ax=ax2, x="Group",y="PSC",hue="Task",
             data=STSData,meanline='True',showfliers = False, **boxplot_kwargs)
                 
gbox2 = sns.stripplot(x="Group",y="PSC",hue="Task", dodge = True, jitter = False,
                    data=STSData, ax=gbox2.axes, edgecolor=None, **stripplot_kwargs)
            
#figure out how to add the gray lines for pairs of points
idx0 = 0; idx1 = 1
locs1 = ax2.get_children()[idx0].get_offsets()
locs2 = ax2.get_children()[idx1].get_offsets()
for i in range(locs1.shape[0]):
    x = [locs1[i, 0], locs2[i, 0]]
    y = [locs1[i, 1], locs2[i, 1]]
    ax2.plot(x, y, color= lcolor, alpha=lalpha, linewidth=lw1, markersize = msize1)
idx0 = 2; idx1 = 3
locs1 = ax2.get_children()[idx0].get_offsets()
locs2 = ax2.get_children()[idx1].get_offsets()
for i in range(locs1.shape[0]):
    x = [locs1[i, 0], locs2[i, 0]]
    y = [locs1[i, 1], locs2[i, 1]]
    ax2.plot(x, y, color= lcolor, alpha=lalpha, linewidth=lw1, markersize = msize1)
gbox2.legend_.remove()
gbox2 = sns.set_style("ticks")
gbox2 = sns.despine(offset = 30, trim = False)
plt.gca().set_ylim(-0.3,1.4,0.3)
ax2.set_yticks(np.arange(-.3, 1.5, 0.3), minor=False)
#ax2.set_xlabel('Expression Task')
ax2.set_ylabel("")
ax2.yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.1f'))
gbox2 = sns.set_context("poster",font_scale=1.2)


############
ax3 = fig.add_subplot(1,3,3)
gbox3 = sns.boxplot(ax=ax3, x="Group",y="PSC",hue="Task",
             data=AMGData,meanline='True',showfliers = False, **boxplot_kwargs)
                 
gbox3 = sns.stripplot(x="Group",y="PSC",hue="Task", dodge = True, jitter = False,
                    data=AMGData, ax=gbox3.axes, edgecolor=None, **stripplot_kwargs)
            
#figure out how to add the gray lines for pairs of points
idx0 = 0; idx1 = 1
locs1 = ax3.get_children()[idx0].get_offsets()
locs2 = ax3.get_children()[idx1].get_offsets()
for i in range(locs1.shape[0]):
    x = [locs1[i, 0], locs2[i, 0]]
    y = [locs1[i, 1], locs2[i, 1]]
    ax3.plot(x, y, color= lcolor, alpha=lalpha, linewidth=lw1, markersize = msize1)
idx0 = 2; idx1 = 3
locs1 = ax3.get_children()[idx0].get_offsets()
locs2 = ax3.get_children()[idx1].get_offsets()
for i in range(locs1.shape[0]):
    x = [locs1[i, 0], locs2[i, 0]]
    y = [locs1[i, 1], locs2[i, 1]]
    ax3.plot(x, y, color= lcolor, alpha=lalpha, linewidth=lw1, markersize = msize1)

gbox3.legend_.remove()
gbox3 = sns.set_style("ticks")
gbox3 = sns.despine(offset = 30, trim = False)
#plt.gca().set_ylim(-0.5,2.5)
#ax3.set_xlabel('Expression Task')
ax3.set_ylabel("")
start, end = plt.gca().get_ylim()
stepsize = 0.2
ax3.yaxis.set_ticks(np.arange(start, end, stepsize))
ax3.yaxis.set_major_formatter(ticker.FormatStrFormatter('%0.1f'))
gbox3 = sns.set_context("poster",font_scale=1.2)


plt.rcParams['hatch.linewidth'] = 4
hatches = [None, "/", None, '/']
colors = ['skyblue', 'skyblue', 'salmon', 'salmon']
a = len(ax1.artists)
#print(a)
for hatch, patch, color in zip(hatches, ax1.artists, colors):
    patch.set_edgecolor('white')
    patch.set_facecolor(color)
    patch.set_hatch(hatch)
for hatch, patch, color in zip(hatches, ax2.artists, colors):
    patch.set_edgecolor('white')
    patch.set_facecolor(color)
    patch.set_hatch(hatch)
for hatch, patch, color in zip(hatches, ax3.artists, colors):
    patch.set_edgecolor('white')
    patch.set_facecolor(color)
    patch.set_hatch(hatch)

plt.show()

```
